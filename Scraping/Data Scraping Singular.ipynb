```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time

# Setup Options
chrome_options = Options()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Gramedia.com has 50 pages
URL = "https://www.gramedia.com/categories/buku/komik?page=10"

# Initiate WebDriver Session
driver = webdriver.Chrome(options=chrome_options)

# Opens The Website in Chrome
driver.get(URL)

# Wait Loading 5s
time.sleep(7)

# Getting the HTML
html = str(driver.page_source)
if (html != ""):
    print("Web HTML Fetched!")
else:
    print("Fetching Failed!")

# Closes Chrome
driver.quit()
```

```python
from bs4 import BeautifulSoup as bsp
import pandas as pd

# Change class into Soup Class
soup = bsp(html, "lxml")

# Print HTML
# print(soup.prettify()[1:1000])

# Based on the HTML, each book and it's data contained in
# "gm-product-list" tag with "card-list ng-star-inserted" class
buku = soup.find_all("gm-product-list", "card-list ng-star-inserted")

# Get All Product Name, Price, and Author and insert it into array
namaBuku = []
hargaBuku = []
namaPenulis = []

for p in buku:
    nama = p.find("div", "list-title").get_text()
    harga = p.find('p', 'formats-price').get_text()
    try:
        penulis = p.find('span', 'list-author ng-star-inserted').get_text()
    except AttributeError: # if the author name is missing
        penulis = ""
    
    # Inserting datum into the array
    # Type String
    namaBuku.append(nama)
    # Type Integer
    hargaBuku.append(int(harga.replace('Rp','').replace(' ','').replace('.','')))
    # Type String
    namaPenulis.append(penulis.strip())

# Creating Dictionary to make Dataframe
prodDict = {'judul':namaBuku, 'harga':hargaBuku, 'penulis':namaPenulis}

# Create dataframe from dictionary
df = pd.DataFrame(prodDict,columns = ['judul','harga','penulis'])

# Show Tables
df
```

```python
from json import loads, dumps
result = df.to_json(orient="table")
parsed = loads(result)
print(dumps(parsed, indent=4))
```
